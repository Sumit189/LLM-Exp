```python
import base64
import os
```

```python
os.environ["HF_TOKEN"] = ''
from byaldi import RAGMultiModalModel
```

```python
RAG = RAGMultiModalModel.from_pretrained("vidore/colpali-v1.2", verbose=1)
```

```python
!mkdir data
```

```python
RAG.index(
    input_path="data/",
    index_name="data",
    store_collection_with_index=True,
    overwrite=True
)
```

```python
RAG.index
```

```python
query = "compared to 2015 onwards which year has lowest number of forest fires"
```

```python
results = RAG.search(query, k=3)
```

```python
results
```

```python
from IPython.display import display, Image
rag_image_bytes = []

for result in results:
  image_bytes = base64.b64decode(result.base64)
  rag_image_bytes.append(result.base64)
  filename = 'image.jpg'
  with open(filename, 'wb') as f:
    f.write(image_bytes)
    display(Image(filename))
```

```python
from openai import AsyncOpenAI
client = AsyncOpenAI(api_key="")
async def get_llm_response(content):
  try:
    response = await client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "user",
                "content": content
            }
        ],
        max_tokens=300
    )

    return response

  except Exception as e:
    print(f"An error occurred: {e}")
    return None
```

```python
user_text = query
content_payload = [
    {
        "type": "text",
        "text": user_text
    }
]
for image_bytes in rag_image_bytes:
  content_payload.append({
      "type": "image_url",
      "image_url": {
          "url": "data:image/jpeg;base64," + image_bytes
          }
      }
  )

response = await get_llm_response(content_payload)
if response:
  print(response.choices[0].message.content)
```
